```python
from numpy import array
from random import uniform
from matplotlib import pyplot
from numpy.linalg import inv
from numpy.random import seed
```

```python
seed(777)
```

```python
sample_size=50
```

```python
X=array([[1*i] for i in range(50)])
```

```python
y=[(10*i)*uniform(0.9,1.1) for i in range(sample_size)]
```

```python
X.T.dot(y)
```

```python
pyplot.scatter(X,y)
```

```python
beta=inv(X.T.dot(X)).dot(X.T).dot(y)
beta
```

```python
y_hat=X.dot(beta)
y_hat
```

```python
pyplot.scatter(X,y)
pyplot.plot(X,y_hat,color="red")
```

```python
y=[(10*i)*uniform(0.9,1.1) + 100 for i in range(sample_size)]
beta=inv(X.T.dot(X)).dot(X.T).dot(y)
y_hat=X.dot(beta)
pyplot.scatter(X,y)
pyplot.plot(X,y_hat,color="red")
```

```python
X=array([[1*i,1] for i in range(50)]) ## CUIDAR EL ORDEN EN QUE SE AGREGA LA COLUMNA 1... 
#COMO AQU SE AGREG DESPUS, LA ESTIMACIN DE LAS BETAS NOS DAR PRIMERO BETA_1 Y LUEGO BETA_0
y=[(10*i)*uniform(0.9,1.1) + 100 for i in range(sample_size)]
beta=inv(X.T.dot(X)).dot(X.T).dot(y)
y_hat=X.dot(beta)
pyplot.scatter(X[:,0],y)
pyplot.plot(X[:,0],y_hat,color="red")
```

```python
X=array([[1*i,1] for i in range(50)]) ## CUIDAR EL ORDEN EN QUE SE AGREGA LA COLUMNA 1... 
#COMO AQU SE AGREG DESPUS, LA ESTIMACIN DE LAS BETAS NOS DAR PRIMERO BETA_1 Y LUEGO BETA_0
y=[((10*i)*uniform(0.9,1.1))**2 + 100 for i in range(sample_size)] #
beta=inv(X.T.dot(X)).dot(X.T).dot(y)
y_hat=X.dot(beta)
pyplot.scatter(X[:,0],y)
pyplot.plot(X[:,0],y_hat,color="red")
```

```python
X=array([[1*i,1,i**2] for i in range(50)]) ## CUIDAR EL ORDEN EN QUE SE AGREGA LA COLUMNA 1... 
#COMO AQU SE AGREG DESPUS, LA ESTIMACIN DE LAS BETAS NOS DAR PRIMERO BETA_1 Y LUEGO BETA_0
y=[((10*i)*uniform(0.9,1.1))**2 + 100 for i in range(sample_size)]
beta=inv(X.T.dot(X)).dot(X.T).dot(y)
y_hat=X.dot(beta)
pyplot.scatter(X[:,0],y)
pyplot.plot(X[:,0],y_hat,color="red")
```
